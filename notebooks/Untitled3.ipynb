```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
rng = np.random.default_rng(seed=0)
x = rng.random((1000, 1))
y = 1 + 5 * x + rng.random((1000, 1))

# Объединяем массивы горизонтально (в два столбца)
# .flatten() нужен, чтобы превратить двумерный массив (1000, 1) в плоский список
data = {"X": x.flatten(), "Y": y.flatten()}
df = pd.DataFrame(data)
df.head()
```

```python
plt.scatter(df['X'], df['Y'], s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show();
```

```python
train, test = train_test_split(df, test_size=0.3, random_state=42)
```

```python
train.shape
```

```python
test.shape
```

```python
plt.scatter(train['X'], train['Y'], s=10,c='b')
plt.scatter(test['X'], test['Y'], s=10,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show();
```

```python
def model(x):
    y_model = 1+5*x
    return y_model

train['y_model'] = train['X'].map(lambda x: model(x))
test['y_model'] = test['X'].map(lambda x: model(x))
```

```python
train.head()
```

```python
plt.plot(train['X'], train['y_model'], linewidth=4, c='g')
plt.scatter(train['X'], train['Y'], s=10,c='b')
plt.scatter(test['X'], test['Y'], s=10,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show();
```

```python
print('MAE train', mean_absolute_error(train['Y'], train['y_model']))
print('MAE test', mean_absolute_error(test['Y'], test['y_model']))
```
