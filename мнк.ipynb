```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([1, 1, 4, 9, 0])
```

```python
plt.scatter(x, y, s=30)
plt.plot(x, y1, 'g')
plt.plot(x, y2, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show();
```

```python
err1 = sum((y1-y)**2)
err2 = sum((y2-y)**2)

print(f'ошибка err1: {err1}')
print(f'ошибка err2: {err2}')
```
